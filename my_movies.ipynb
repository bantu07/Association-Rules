```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C://Users//Dell//Downloads//Excel R//Assignments//Machine Learning//Unsupervised//Association rules//my_movies.csv",header=None)
df
```

```python
transact = []
for i in range(0,9):
    transact.append([str(df.values[i,j]) for j in range(0,14)])
    while 'nan' in transact[i]:
        transact[i].remove('nan')
```

```python
from apyori import apriori
rules = apriori(transact,min_support = 0.0045,min_confidence=0.20,min_lift=4,min_length=2)
rules
```

```python
results = list(rules)
print(len(results))
```

```python
print(results[0])
```

```python
for item in results:
    pair = item[0]
    items = [x for x in pair]

    print("Rule: " + str(list(item.ordered_statistics[0].items_base)) + " -> " + str(list(item.ordered_statistics[0].items_add)))
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
